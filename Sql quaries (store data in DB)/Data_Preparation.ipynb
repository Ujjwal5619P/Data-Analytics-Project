```python
import pandas as pd
import numpy as np
import os
import re

# Load the dataset
folder = 'c:/Users/shubh/Desktop/Tableau_Project'
file_path = os.path.join(folder, 'Heart_new2 (1).csv')
df = pd.read_csv(file_path)

print('Initial Shape:', df.shape)
print('Missing values:', df.isnull().sum().sum())
print('Duplicates:', df.duplicated().sum())

# Display the first few rows
df.head()
```

```python
def split_camel_case(s):
    if s == 'BMI': return 'BMI'
    return re.sub(r'(?<!^)(?=[A-Z])', ' ', s)

new_cols = {col: split_camel_case(col) for col in df.columns}
df.rename(columns=new_cols, inplace=True)

print('New Columns:', df.columns.tolist())
```

```python
def get_bmi_category(bmi):
    if bmi < 18.5: return 'Underweight'
    elif bmi < 25: return 'Normal Weight'
    elif bmi < 30: return 'Overweight'
    else: return 'Obese'

df['BMI Category'] = df['BMI'].apply(get_bmi_category)
df['BMI Category'].value_counts()
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)

print('Final Shape:', df.shape)
df.head()
```

```python
out_path = os.path.join(folder, 'Heart_Disease_Tableau_Ready.csv')
df.to_csv(out_path, index=False)
print('Saved prepared dataset to:', out_path)
```
